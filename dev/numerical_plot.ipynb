```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0, 30, 100)
y1 = (2 * x) / (x + 2)
y2 = (2 * x) / (0.5 * x + 2)

plt.plot(x, y1, label='Assume d_v^D = 0')
plt.plot(x, y2, label='Assume d_v^D = 0.5 d_v')
plt.xlabel('d_v')
plt.ylabel('Cost Utility')
plt.title('Cost Utility vs Degree in Linear Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0, 210, 100)
y1 = (2 * x**(1/2)) / (x + 2)
y2 = (2 * x**(1/2)) / (0.5 * x + 2)

plt.plot(x, y1, label='Assume d_v^D = 0')
plt.plot(x, y2, label='Assume d_v^D = 0.5 d_v')
plt.xlabel('d_v')
plt.ylabel('Cost Utility')
plt.title('Cost Utility vs Degree in Square Root Price')
plt.grid(True)
plt.legend()
plt.show()
```
